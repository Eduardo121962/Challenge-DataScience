### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja1.columns
```

#1. Análise do faturamento


```python
loja1.columns

```

```python
categorias =

colunas = ['LOJA1','LOJA2','LOJA3','LOJA4']
dados = []

listaprecoloja1 = list(loja1['Preço'].tolist())
total_loja1 = round(sum(listaprecoloja1),2)
print(f'LOJA1 R$ {total_loja1}\n')
dados.append(total_loja1)

listaprecoloja2 = list(loja2['Preço'].tolist())
total_loja2 = round(sum(listaprecoloja2),2)
print('LOJA2 R$',total_loja2,'\n')
dados.append(total_loja2)

listaprecoloja3 = list(loja3['Preço'].tolist())
total_loja3 = round(sum(listaprecoloja3),2)
print('LOJA3 R$',total_loja3,'\n')
dados.append(total_loja3)

listaprecoloja4 = list(loja4['Preço'].tolist())
total_loja4 = round(sum(listaprecoloja4),2)
print('LOJA4 R$',total_loja4,'\n')
dados.append(total_loja4)

faturamento_total = total_loja1 + total_loja2 + total_loja3 + total_loja4
print("Faturamento total = R$",faturamento_total)

plt.pie(dados,labels = colunas,autopct='%1.1f%%')
plt.show()
```

```python
qtde_vend_Loja1 = loja1.groupby('Categoria do Produto')['Categoria do Produto'].count()
categorias = qtde_vend_Loja1.index.tolist()
print(categorias)
```

# 2. Vendas por Categoria


```python
qtde_vend_Loja1 = loja1.groupby('Categoria do Produto')['Categoria do Produto'].count()
print('Loja1\n Quantidade vendida por categoria')
print(qtde_vend_Loja1)
vendasCatLoja1 = loja1.groupby('Categoria do Produto')['Preço'].sum()
print('\n \n Total de vendas por categoria\n',vendasCatLoja1,'\n')

plt.pie(qtde_vend_Loja1, labels = categorias, autopct='%1.1f%%')
plt.show()
plt.pie(vendasCatLoja1, labels = categorias, autopct='%1.1f%%')
plt.show()

print('--------------------------------------------------------------------')

qtde_vend_Loja2 = loja2.groupby('Categoria do Produto')['Categoria do Produto'].count()
print('Loja2\n Quantidade vendida por categoria')
print(qtde_vend_Loja2)
vendasCatLoja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
print('\n \n Total de vendas por categoria\n',vendasCatLoja1,'\n')

plt.pie(qtde_vend_Loja2, labels = categorias, autopct='%1.1f%%')
plt.show()
plt.pie(vendasCatLoja2, labels = categorias, autopct='%1.1f%%')
plt.show()

print('--------------------------------------------------------------------')

qtde_vend_Loja3 = loja3.groupby('Categoria do Produto')['Categoria do Produto'].count()
print('Loja3\n Quantidade vendida por categoria')
print(qtde_vend_Loja3)
vendasCatLoja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
print('\n \n Total de vendas por categoria\n',vendasCatLoja1,'\n')

plt.pie(qtde_vend_Loja3, labels = categorias, autopct='%1.1f%%')
plt.show()
plt.pie(vendasCatLoja3, labels = categorias, autopct='%1.1f%%')
plt.show()

print('--------------------------------------------------------------------')

qtde_vend_Loja4 = loja4.groupby('Categoria do Produto')['Categoria do Produto'].count()
print('Loja1\n Quantidade vendida por categoria')
print(qtde_vend_Loja4)
vendasCatLoja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()
print('\n \n Total de vendas por categoria\n',vendasCatLoja1,'\n')

plt.pie(qtde_vend_Loja2, labels = categorias, autopct='%1.1f%%')
plt.show()
plt.pie(vendasCatLoja4, labels = categorias, autopct='%1.1f%%')
plt.show()

print('--------------------------------------------------------------------')
```

```python
lista_loja1 = []
lista_loja2 = []
lista_loja3 = []
lista_loja4 = []
# agrupando por categoria e somando os preços para cada categoria
loja1_agrupado = loja1.groupby('Categoria do Produto')['Preço'].sum().reset_index()
loja2_agrupado = loja2.groupby('Categoria do Produto')['Preço'].sum().reset_index()
loja3_agrupado = loja3.groupby('Categoria do Produto')['Preço'].sum().reset_index()
loja4_agrupado = loja4.groupby('Categoria do Produto')['Preço'].sum().reset_index()

#criando uma lista com os preços por loja
for i in range(8):
  lista_loja1.append(float(loja1_agrupado['Preço'][i]))
  lista_loja2.append(float(loja2_agrupado['Preço'][i]))
  lista_loja3.append(float(loja3_agrupado['Preço'][i]))
  lista_loja4.append(float(loja4_agrupado['Preço'][i]))

plt.barh(loja1_agrupado['Categoria do Produto'],loja1_agrupado['Preço'], label = "Loja1")
plt.legend()
plt.barh(loja4_agrupado['Categoria do Produto'],loja4_agrupado['Preço'], label = "Loja4")
plt.legend()
plt.show()

plt.barh(loja2_agrupado['Categoria do Produto'],loja2_agrupado['Preço'], label = "Loja2")
plt.legend()
plt.barh(loja4_agrupado['Categoria do Produto'],loja4_agrupado['Preço'], label = "Loja4")
plt.legend()
plt.show()

plt.barh(loja3_agrupado['Categoria do Produto'],loja3_agrupado['Preço'], label = "Loja3")
plt.legend()
plt.barh(loja4_agrupado['Categoria do Produto'],loja4_agrupado['Preço'], label = "Loja4")
plt.legend()
plt.show()
```
